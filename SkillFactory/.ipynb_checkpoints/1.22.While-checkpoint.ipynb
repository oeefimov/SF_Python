```python
#Предположим, что мы положили деньги в банк под 6 процентов годовых. Через сколько лет деньги удвоятся? 
#Такую задачу сложно решить с помощью цикла for, однако с ней нам поможет  цикл while. 
#Цикл while («пока») позволяет выполнить одну и ту же последовательность действий, пока проверяемое условие истинно. 
#Условие записывается до тела цикла и проверяется до выполнения тела цикла. 
#Как правило, цикл while используется, когда невозможно определить точное значение количества проходов исполнения цикла.

#Вернёмся к задаче и решим её с помощью цикла while:
#Каждую итерацию цикла проверяется условие того, что money < 2. 
#Если условие выполняется, то выполняется и тело цикла, в котором к year прибавляется 1, а money умножается на 1.06 (+6%). 
#Как только мы получили удвоение денег, то есть money >= 2, цикл перестаёт выполняться, 
#и выполняется блок кода после else, который выводит нужное количество лет. 

money = 1
year = 0
yearly_multiplier = 1.06

while money < 2:
    money = money*yearly_multiplier
    year += 1
else:
    print('Деньги удвоятся через', year, 'лет')

#Output:
#Деньги удвоятся через 12 лет 
```

```python
#Задание 1

#Клиент банка открыл вклад под 10% годовых и положил на этот счёт Y рублей. 
#Напишите программу, позволяющую определить, через сколько лет клиент будет иметь на счету сумму не менее Z рублей. 
#На вход программе идут два числа — Y, Z.
#Для проверки используйте Y = 170000, Z = 1000000. 
#В результате работы программа должна вывести на экран число лет, которое потребуется для накопления требуемой суммы. 
#Ответ должен быть выведен на экран в виде целого числа без какого-либо дополнительного поясняющего текста.
#Внимание: в процессе выполнения программы изменяйте переменную Y.

Y = 170000
Z = 1000000
year=0
while Y<Z:
    Y=Y*1.1
    year+=1
else:
    print(year)
```

```python
#Задание 2

#Пассажирский лифт имеет ограничение на перевозку: не более 400 кг единовременно. 
#Пусть на вход программе поступает вес (weight) входящих в лифт людей, а в случае, когда возникает перевес, 
#программа выдаёт предупреждение: "Перевес N кг". 
## КРЕТИНЫ. weight - вес КАЖДОГО ИЗ входящих в лифт людей
#Например, при перевесе в 15 кг должно быть выведено сообщение: "Перевес 15 кг". 
#Обратите внимание, что в эталонной фразе не используются знаки препинания.
#Для проверки используйте weight = 77 и не меняйте его по ходу цикла.

weight = 77
total=0
while total<400:
    total+=weight
else:
    print('Перевес',total-400,'кг')
```
