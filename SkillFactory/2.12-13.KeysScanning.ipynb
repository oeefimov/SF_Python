```python
2.12 Перебор элементов словаря. Постановка задачи

Очень часто возникает задача перебора элементов словаря. 
Рассмотрим такую задачу: в корпоративном справочнике имеется следующая структура должностей
employee_base = {'Мария Никитина': 'менеджер', 'Егор Савичев': 'разработчик', 'Александр Пахомов': 'дизайнер', 
                 'Алина Егорова': 'разработчик', 'Руслан Башаров': 'верстальщик'}

Необходимо отфильтровать сотрудников в должности разработчика. Чтобы получить значение, нам нужно отфильтровать значения. 
И есть те ключи, для которых значение равно 'разработчик'.

Для перебора ключей и значений словаря есть три основных способа:
 - ключи, # см. метод .keys()
 - значения, # см. метод .values()
 - ключи и значения одновременно.

Мы получаем простой цикл, в каждом шаге которого есть переменные значения одного из этих вариантов.


2.13 Перебор ключей словаря


Для перебора ключей используется простой синтаксис (название переменной цикла employee можно выбрать любым):

for employee in employee_base: #т.е. здесь мы "объявляем" переменную employee и присваиваем ей значения ключей
    print(employee)
#перебор ключей вариант 1
    
Также проход по ключам можно осуществить с помощью метода keys. Результат будет аналогичным:

for employee in employee_base.keys():
    print(employee)
#перебор ключей вариант 1а
    
Теперь, зная в каждом шаге цикла ключ, можно вывести соответствующее значение (у нас это должность сотрудника):

for employee in employee_base:
    print(employee, employee_base[employee])
#результат прохода по ключам

Зная значение, можно наложить фильтр (напомним, мы хотим получить имена разработчиков) и решить нашу задачу:

for employee in employee_base:
    if employee_base[employee] == 'разработчик':
        print(employee)
#Результат способом обхода ключей ##Чем "обход" отличается от "прохода"?

Получилось довольно громоздко. Для нашей задачи есть немного более изящный способ решения. 
Но несмотря на это, проход по ключам словаря очень удобен во многих задачах. 
Давайте посмотрим другие способы обхода словаря.
```

```python
#перебор ключей вариант 1
employee_base = {'Мария Никитина': 'менеджер', 'Егор Савичев': 'разработчик', 'Александр Пахомов': 'дизайнер', 
                 'Алина Егорова': 'разработчик', 'Руслан Башаров': 'верстальщик'}
for employee in employee_base:
    print(employee)
```

```python
#Фильтрация результатов жеребьевки

#Отфильтруйте словарь draw_dict, оставив в нем только страны группы A. 
#Отфильтрованный результат запишите в новый словарь draw_new

draw_dict = {
	'Россия': 'A',
	'Португалия': 'B',
	'Франция': 'C',
	'Дания': 'C',
	'Египет': 'A'
}

#Подсказка:
#Для добавления в словарь нового элемента в коде нужно указать имя словаря, 
#далее в квадратных скобках указать новый ключ (в виде строки или имени переменной), 
#поставить знак " = " и после этого указать значение, которое необходимо установить для данного ключа.
```

```python
draw_dict = {
	'Россия': 'A',
	'Португалия': 'B',
	'Франция': 'C',
	'Дания': 'C',
	'Египет': 'A'
}
for team in draw_dict:
    if draw_dict[team] == 'A':
        print(team)
```

```python
#Тоже оценили как правильный, хотя в подсказке давался следующий вариант
draw_dict = {
	'Россия': 'A',
	'Португалия': 'B',
	'Франция': 'C',
	'Дания': 'C',
	'Египет': 'A'
}
draw_new={}
for team in draw_dict:
    if draw_dict[team] == 'A':
        draw_new.setdefault(team, 'A')
print(draw_new)
```
